```python
def cdl_processing(cdl_metadata, qgis_csv):
    metadata = pd.read_csv(cdl_metadata)
    qgis_df = pd.read_csv(qgis_csv)
    metadata["cdl_number"] = "VALUE_" + metadata["cdl_number"].astype("str")
    qgis_df.rename(
        columns=metadata.set_index("cdl_number")["cdl_classification"], inplace=True
    )
    qgis_df = qgis_df.drop(
        [
            "descriptio",
            "timestamp",
            "begin",
            "end",
            "altitudeMo",
            "tessellate",
            "extrude",
            "visibility",
            "drawOrder",
            "icon",
        ],
        axis=1,
    )
    qgis_df["Total"] = qgis_df.sum(axis=1, numeric_only=True)
```
