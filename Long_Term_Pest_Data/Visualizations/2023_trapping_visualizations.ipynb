```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../IPM_Dashboard/trapping_data_2023.csv')

df
```

```python
farm_df = df.loc[df['Farm'] == 'AG']

fig, axs = plt.subplots(5, figsize=(8, 20))
fig.suptitle('Sweet Corn Pheromone Trap Catches, 2023', fontsize=16, y=0.91)


axs[0].plot(farm_df['Date'], farm_df['ECB-E'])
axs[0].set_ylim([0,50])
axs[0].set_title('ECB-E')

axs[1].plot(farm_df['Date'], farm_df['ECB-Z'])
axs[1].set_ylim([0,50])
axs[1].set_title('ECB-Z')

axs[2].plot(farm_df['Date'], farm_df['CEW'])
axs[2].set_ylim([0,50])
axs[2].set_title('CEW')

axs[3].plot(farm_df['Date'], farm_df['FAW'])
axs[3].set_ylim([0,50])
axs[3].set_title('FAW')

axs[4].plot(farm_df['Date'], farm_df['WBC'])
axs[4].set_ylim([0,50])
axs[4].set_title('WBC')


```
