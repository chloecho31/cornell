```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../IPM_Dashboard/trapping_data_2023.csv')
df
```

```python
farms = pd.unique(df['Farm'])
farms = np.array(['REM', 'SS', 'MAN', 'FREE', 'HAT', 'MUS', 'FISH', 'RUSH',
       'AG', 'KIR', 'OLES', 'WH',
       'OLES_2', 'RUSH_2', 'KIR_2', 'MAN_2'])
farms
```

```python
for farm in farms:     

    farm_df = df.loc[df['Farm'] == farm]

    fig, axs = plt.subplots(5, figsize=(8, 30))
    fig.suptitle('Sweet Corn Pheromone Trap Catches, 2023 - ' + farm, fontsize=16, y=0.9)

    axs[0].plot(farm_df['Date'], farm_df['ECB-E'])
    axs[0].set_ylim([0,45])
    axs[0].set_title('ECB-E')
    axs[0].tick_params(labelrotation=30)

    axs[1].plot(farm_df['Date'], farm_df['ECB-Z'])
    axs[1].set_ylim([0,45])
    axs[1].set_title('ECB-Z')
    axs[1].tick_params(labelrotation=30)

    axs[2].plot(farm_df['Date'], farm_df['CEW'])
    axs[2].set_ylim([0,45])
    axs[2].set_title('CEW')
    axs[2].tick_params(labelrotation=30)

    axs[3].plot(farm_df['Date'], farm_df['FAW'])
    axs[3].set_ylim([0,45])
    axs[3].set_title('FAW')
    axs[3].tick_params(labelrotation=30)

    axs[4].plot(farm_df['Date'], farm_df['WBC'])
    axs[4].set_ylim([0,45])
    axs[4].set_title('WBC')
    axs[4].tick_params(labelrotation=30)

    plt.savefig(farm + '_pest_count.png')
```

```python
large_counts = ['CURC', 'FISH_2', 'HAR', 'SEN', 'THOR', 'ZIT']

for farm in large_counts:     

    farm_df = df.loc[df['Farm'] == farm]

    fig, axs = plt.subplots(5, figsize=(8, 30))
    fig.suptitle('Sweet Corn Pheromone Trap Catches, 2023 - ' + farm, fontsize=16, y=0.9)

    axs[0].plot(farm_df['Date'], farm_df['ECB-E'])
    axs[0].set_ylim([0,50])
    axs[0].set_title('ECB-E')
    axs[0].tick_params(labelrotation=30)

    axs[1].plot(farm_df['Date'], farm_df['ECB-Z'])
    axs[1].set_ylim([0,75])
    axs[1].set_title('ECB-Z')
    axs[1].tick_params(labelrotation=30)

    axs[2].plot(farm_df['Date'], farm_df['CEW'])
    axs[2].set_ylim([0,75])
    axs[2].set_title('CEW')
    axs[2].tick_params(labelrotation=30)

    axs[3].plot(farm_df['Date'], farm_df['FAW'])
    axs[3].set_ylim([0,50])
    axs[3].set_title('FAW')
    axs[3].tick_params(labelrotation=30)

    axs[4].plot(farm_df['Date'], farm_df['WBC'])
    axs[4].set_ylim([0,90])
    axs[4].set_title('WBC')
    axs[4].tick_params(labelrotation=30)

    plt.savefig(farm + '_pest_count.png')
```
