```python
import pandas as pd
```

```python
# Read CSV file. 

pest_df = pd.read_csv('NYSIPM_Sweet_Corn_1993_2022.csv')
pest_df
```

```python
# Subset columns from original dataframe. 

pest_df = pest_df[['Site', 
                   'Lat-Decimal', 
                   'Long-Decimal', 
                   'Year', 
                   'Week',
                   'Date',
                   'ECB-E',
                   'ECB-Z',
                   'CEW',
                   'FAW',
                   'WBC']]

pest_df
```

```python
pest_df['Site'] = pest_df['Site'].astype('category')
```

```python
pest_df['Date'] = pd.to_datetime(pest_df['Date'])
pest_df['Day'] = pest_df[ "Date"].dt.dayofyear
```

```python
pest_df
```

```python
pd.melt(pest_df, id_vars = ['Site', 
                            'Lat-Decimal', 
                            'Long-Decimal',
                            'Year', 
                            'Week', 
                            'Date', 
                            'Day'], 
                 value_vars = ['ECB.E', 
                                 'ECB.Z',
                                 'CEW', 
                                 'FAW', 
                                 'WBC'])
```
