```python
wire_mesh_23 = pd.read_csv("../2023/Data/2023_wire_mesh_clean.csv")
wire_mesh_23 = clean_columns(wire_mesh_23)
wire_mesh_23["collection_date"] = pd.to_datetime(wire_mesh_23["collection_date"], format="%m/%d/%y")
wire_mesh_23
```

```python
wire_mesh_23['total_scm'] = wire_mesh_23['total_scm_maggots'] + wire_mesh_23['total_scm_pupae']
wire_mesh_23
```

```python
wire_mesh_23 = wire_mesh_23[['collection_date', 'record_id', 'total_scm']]
wire_mesh_23
```

```python
adults_23 = pd.read_csv("../2023/Data/2023_scm_counts.csv")
adults_23 = clean_columns(adults_23)
adults_23["collection_date"] = pd.to_datetime(adults_23["collection_date"], format="%m/%d/%y")
adults_23["week"] = adults_23["collection_date"].dt.strftime("%U")
#adults_23 = adults_23.drop(['initials', 'notes'], axis=1)
#adults_23 = adults_23.dropna()
adults_23['total_adult_scm'] = adults_23['n_scm_i_m'] + adults_23['n_scm_i_f'] + adults_23['n_scm_o_m'] + adults_23['n_scm_o_f']
adults_23
```

```python
combined_23 = wire_mesh_23.merge(adults_23, left_on=['collection_date', 'record_id'], right_on=['collection_date', 'site']).drop(['site'], axis=1)
combined_23
```

```python
adults_24 = pd.read_csv("Data/2024_scm_counts.csv")
adults_24 = clean_columns(adults_24)
adults_24["date_collected"] = pd.to_datetime(adults_24["date_collected"])
adults_24["week"] = adults_24["date_collected"].dt.strftime("%U")
adults_24 = adults_24.drop(['initials', 'unnamed_12'], axis=1)
adults_24 = adults_24.dropna()
adults_24['total_adult_scm'] = adults_24['n_scm_m_1'] + adults_24['n_scm_m_2'] + adults_24['n_scm_f_1'] + adults_24['n_scm_f_2']
adults_24
```

```python
combined_24 = wire_mesh_24.merge(adults_24, left_on=['collection_date', 'field_id'], right_on=['date_collected', 'field_id'])
combined_24
```

```python
combined_23 = combined_23.rename(columns={"record_id": "field_id"})
combined_23
```

```python
combined_24 = combined_24[['field_id', 'collection_date', 'total_scm', 'total_adult_scm']]
combined_24
```

```python
combined_23 = combined_23[['field_id', 'total_scm', 'collection_date', 'total_adult_scm']]
combined_23
```

```python
all_combined = pd.concat([combined_23, combined_24], ignore_index=True,axis=0)
all_combined.to_csv('all_combined.csv')
```

```python
all_combined
```

```python
plt.rcParams['font.sans-serif'] = ['Avenir']
plt.style.use('default')
plt.figure(figsize=(10,8))

plt.rcParams.update({'font.size': 24})

sns.regplot(
    data=all_combined, x="total_adult_scm", y="total_scm",
    scatter=True, truncate=False, order=1, color="black",
)

plt.xlabel('Adults')
plt.ylabel('Larvae')
plt.title('$\it{D. platura}$ larvae vs. adults')
plt.savefig('combined_scm.png')
plt.show()

```
