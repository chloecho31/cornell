```python
import pandas as pd
import numpy as np
from skimpy import clean_columns
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import savefig
from scipy import stats
import scikit_posthocs as sp


```

```python
wire_mesh_24 = pd.read_csv("Data/scm_wire_mesh_2024.csv")
wire_mesh_24 = clean_columns(wire_mesh_24)
wire_mesh_24
```

```python
wire_mesh_24 = wire_mesh_24.loc[wire_mesh_24['project'] == 'R']
wire_mesh_24
```

```python
wire_mesh_24["collection_date"] = pd.to_datetime(
    wire_mesh_24["collection_date"], format='%m/%d/%y'
)
```

```python
wire_mesh_24['total_scm'] = wire_mesh_24['scm_larvae_sum'] + wire_mesh_24['scm_pupae_sum']
wire_mesh_24['total_phorids'] = wire_mesh_24['phorid_larvae_sum'] + wire_mesh_24['phorid_pupae_sum']

wire_mesh_24['week'] = wire_mesh_24['collection_date'].dt.isocalendar().week.astype(int)
wire_mesh_24['week_before'] = wire_mesh_24['week'] - 1
wire_mesh_24['two_weeks_before'] = wire_mesh_24['week'] - 2

wire_mesh_24
```

```python
wire_mesh_24 = wire_mesh_24[['field_id', 'collection_date', 'total_scm', 'number_seeds', 'scm_seeds', 'seed_type', 'wireworms', 'total_phorids', 'scm_larvae_sum', 'scm_pupae_sum', 'week', 'week_before', 'two_weeks_before']]
wire_mesh_24.to_csv('Data/2024_wire_mesh_clean.csv')
```

```python
wire_mesh_23 = pd.read_csv("../2023/Data/2023_wire_mesh_clean.csv")
wire_mesh_23 = clean_columns(wire_mesh_23)
wire_mesh_23["collection_date"] = pd.to_datetime(wire_mesh_23["collection_date"], format="%m/%d/%y")
wire_mesh_23
```

```python
wire_mesh_23['total_scm'] = wire_mesh_23['total_scm_maggots'] + wire_mesh_23['total_scm_pupae']
wire_mesh_23['week'] = wire_mesh_23['collection_date'].dt.isocalendar().week.astype(int)
wire_mesh_23['week_before'] = wire_mesh_23['week'] - 1
wire_mesh_23['two_weeks_before'] = wire_mesh_23['week'] - 2
wire_mesh_23
```

```python
adults_23 = pd.read_csv("../2023/Data/2023_scm_counts.csv")
adults_23 = clean_columns(adults_23)
adults_23['n_scm_i_m'] = pd.to_numeric(adults_23['n_scm_i_m'], errors='coerce')
adults_23['n_scm_i_f'] = pd.to_numeric(adults_23['n_scm_i_f'], errors='coerce')
adults_23['n_scm_o_m'] = pd.to_numeric(adults_23['n_scm_o_m'], errors='coerce')
adults_23['n_scm_o_f'] = pd.to_numeric(adults_23['n_scm_o_f'], errors='coerce')
adults_23['n_d_florilega_i'] = pd.to_numeric(adults_23['n_d_florilega_i'], errors='coerce')
adults_23['n_d_florilega_o'] = pd.to_numeric(adults_23['n_d_florilega_o'], errors='coerce')
adults_23["date"] = pd.to_datetime(adults_23["date"], format="%m/%d/%y")
adults_23["week"] = adults_23["date"].dt.strftime("%U").astype(int)
#adults_23 = adults_23.drop(['initials', 'notes'], axis=1)
#adults_23 = adults_23.dropna()
adults_23['total_adult_scm'] = adults_23['n_scm_i_m'] + adults_23['n_scm_i_f'] + adults_23['n_scm_o_m'] + adults_23['n_scm_o_f']
adults_23['total_flies'] = adults_23['total_adult_scm'] + adults_23['n_d_florilega_i'] + adults_23['n_d_florilega_o']
adults_23
```

```python
combined_23 = wire_mesh_23.merge(adults_23, left_on=['collection_date', 'record_id'], right_on=['date', 'site']).drop(['site'], axis=1)
combined_23
```

```python
adults_24 = pd.read_csv("Data/2024_scm_counts.csv")
adults_24 = clean_columns(adults_24)
adults_24["date_collected"] = pd.to_datetime(adults_24["date_collected"])
adults_24["week"] = adults_24["date_collected"].dt.strftime("%U").astype(int)
adults_24 = adults_24.drop(['initials'], axis=1)
adults_24 = adults_24.dropna()
adults_24['total_adult_scm'] = adults_24['n_scm_m_1'] + adults_24['n_scm_m_2'] + adults_24['n_scm_f_1'] + adults_24['n_scm_f_2']
adults_24['total_flies'] = adults_24['total_adult_scm'] + adults_24['n_d_florilega_1'] + adults_24['n_d_florilega_2']
adults_24
```

```python
plt.rcParams['font.sans-serif'] = ['Avenir']
plt.rcParams.update({'font.size': 24})
plt.figure(figsize=(12,8))

df = all_combined.groupby('seed_type')['total_scm'].mean().reset_index()
se_df = all_combined.groupby('seed_type')['total_scm'].sem().reset_index()
custom_dict = {'Corn': 0, 'Soybean': 1, 'Lima Bean': 2}
df = df.sort_values(by=['seed_type'], key=lambda x: x.map(custom_dict))
se_df = se_df.sort_values(by=['seed_type'], key=lambda x: x.map(custom_dict))

ax = df.plot.bar(color="#0e7482", rot=0, legend=False)
ax.set_ylim(0, 5)
ax.set_xticklabels(df.seed_type)
#plt.bar_label(ax.containers[0], fmt='%.3f')
plt.errorbar(x=df["seed_type"], y=df["total_scm"], yerr=se_df['total_scm'], ecolor='black', fmt='None')


#plt.xlabel('Seed Type')
plt.xlabel(' ')
#lt.ylabel('$\it{D. platura}$ larvae')
#plt.title('Average Number of $\it{D. platura}$ Larvae per Trap')
#plt.savefig('Figures/larvae_per_trap.png')
plt.show()
```

```python
lima = all_combined.loc[all_combined['seed_type'] == 'Lima Bean']['total_scm']
corn = all_combined.loc[all_combined['seed_type'] == 'Corn']['total_scm']
soy = all_combined.loc[all_combined['seed_type'] == 'Soybean']['total_scm']
stats.kruskal(lima, corn, soy)
```

```python
data = [lima, corn, soy]
p_values = sp.posthoc_dunn(data, p_adjust='holm')

print(p_values)
```
