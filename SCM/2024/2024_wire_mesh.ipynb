```python
import pandas as pd
import numpy as np
from skimpy import clean_columns
```

```python
wire_mesh = pd.read_csv("Data/scm_wire_mesh_2024.csv")
wire_mesh = clean_columns(wire_mesh)
wire_mesh
```

```python
wire_mesh = wire_mesh.loc[wire_mesh['project'] == 'R']
wire_mesh
```

```python
len(np.unique(wire_mesh['field_id']))
```

```python
wire_mesh["collection_date"] = pd.to_datetime(
    wire_mesh["collection_date"], format='%m/%d/%y'
)
```

```python
wire_mesh['total_scm'] = wire_mesh['scm_larvae_sum'] + wire_mesh['scm_pupae_sum']
wire_mesh
```

```python
wire_mesh.groupby(['seed_type'])['total_scm'].sum().plot(kind='bar')
```

```python
wire_mesh.groupby(['seed_type'])['total_scm'].mean().plot(kind='bar')
```

```python
wire_mesh.plot.scatter(x='collection_date', y='total_scm')
```
