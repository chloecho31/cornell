```python
import pandas as pd
import numpy as np
from skimpy import clean_columns
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date
from datetime import datetime
import scipy.stats as stats
from datetime import timedelta
```

```python
temp = pd.read_csv('Data/temps.csv')
temp
```

```python
planting_dates = pd.read_csv('Data/planting_dates.csv')
planting_dates
```

```python
merged = temp.merge(planting_dates, on = 'Field ID')
merged['Date'] = pd.to_datetime(merged["Date"], format='mixed')

merged['Planting date'] = pd.to_datetime(merged["Planting date"], format='mixed')

merged
```

```python
merged['Diff'] = merged['Planting date'] - merged['Date'] 
```

```python
merged = merged.loc[(merged['Diff'] <= '7 days') & (merged['Diff'] >= '-7 days')]
merged
```

```python
two_week_avg = pd.DataFrame(merged.groupby('Field ID')[['tmean (degrees F)', 'tmin (degrees F)', 'tmax (degrees F)']].mean())
two_week_avg.reset_index()
```

```python
two_week_avg.mean()
```

```python
temp_incidence = pd.read_csv('Data/temps_incidence.csv')
temp_incidence
```

```python
planting_dates_incidence = pd.read_csv('Data/planting_dates_incidence.csv')
planting_dates_incidence
```
