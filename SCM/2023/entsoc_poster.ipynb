```python
import pandas as pd
import numpy as np
from skimpy import clean_columns
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import savefig
from scipy import stats
```

Adults vs. larvae scatterplot in 2024_wire_mesh.ipynb.

```python
adult_features = pd.read_csv('Data/2023_adult_features.csv')
adult_features
```

```python
plt.rcParams['font.sans-serif'] = ['Avenir']
plt.style.use('default')
plt.figure(figsize=(10,8))

plt.rcParams.update({'font.size': 25})

sns.regplot(
    data=adult_features, x="ave_temp_f_1_week", y="total_scm",
    scatter=True, truncate=False, order=1, color="black",
)

plt.xlabel('Weekly Average Temperature (Â°F)')
plt.ylabel('Adults')
plt.title('$\it{D. platura}$ adults vs. temperature')
plt.savefig('Figures/weekly_temp.png')
plt.show()
```

```python
stats.pearsonr(adult_features['ave_temp_f_1_week'], adult_features['total_scm'])
```

```python
adult_features['cover_crop'] = adult_features['cover_crop'].replace('Partial', 'Yes')
```

```python
plt.rcParams['font.sans-serif'] = ['Avenir']
plt.style.use('default')
plt.figure(figsize=(10,8))

plt.rcParams.update({'font.size': 25})

sns.barplot(
    data=adult_features, x="cover_crop", y="total_scm",color="black",
)

plt.xlabel('Cover Crop (%)')
plt.ylabel('Adults')
plt.title('$\it{D. platura}$ adults vs. cover crop')
plt.savefig('Figures/cover_crop.png')
plt.show()
```

```python
yes_cc = adult_features.loc[adult_features['cover_crop'] == 'Yes']['total_scm']
no_cc = adult_features.loc[adult_features['cover_crop'] == 'No']['total_scm']
```

```python
stats.ttest_ind(yes_cc, no_cc)
```
